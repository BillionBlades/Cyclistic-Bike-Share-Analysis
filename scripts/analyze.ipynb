```python
# Mapping angka ke nama hari
day_name_map = {1: 'Sunday', 2: 'Monday', 3: 'Tuesday', 4: 'Wednesday', 5: 'Thursday', 6: 'Friday', 7: 'Saturday'}

# 1. Summary Statistics (Tetap perlu untuk laporan performa durasi)
# Kita gunakan reset_index() agar 'member_casual' menjadi kolom, bukan index.
summary_stats = df_clean.groupby('member_casual')['ride_length'].agg(['mean', 'median', 'max', 'min']).reset_index()

# 2. Agregasi Jumlah Perjalanan (Format Long untuk Tableau)
rides_by_day = df_clean.groupby(['day_of_week', 'member_casual']).size().reset_index(name='total_rides')
rides_by_day['day_of_week'] = rides_by_day['day_of_week'].map(day_name_map)

# 3. Agregasi Durasi Rata-rata (Format Long untuk Tableau)
avg_duration_day = df_clean.groupby(['day_of_week', 'member_casual'])['ride_length'].mean().reset_index(name='avg_duration')
avg_duration_day['day_of_week'] = avg_duration_day['day_of_week'].map(day_name_map)

# 4. Simpan semua ke CSV
summary_stats.to_csv('summary_statistics.csv', index=False)
rides_by_day.to_csv('rides_by_day.csv', index=False)
avg_duration_day.to_csv('duration_by_day.csv', index=False)

print("Semua file telah berhasil dibuat!")
```
