```python
# 3. Ambil semua file CSV dalam folder
path = '/content/drive/MyDrive/Cyclistic-Bike-Share-Analysis/cyclistic_final_clean.csv'
df_clean = pd.read_csv(path)
```

# Analyze

```python
df_clean
```

```python
# Analisis perbandingan Member vs Casual
summary_stats = df_clean.groupby('member_casual')['ride_length'].agg(['mean', 'median', 'max', 'min'])

# Analisis jumlah perjalanan per hari
rides_per_day = df_clean.groupby(['member_casual', 'day_of_week']).size().reset_index(name='total_rides')
rides_per_day = rides_per_day.pivot(index='day_of_week', columns='member_casual', values='total_rides')

# Analisis Rata-rata Durasi per Hari
avg_duration_day = df_clean.groupby(['member_casual', 'day_of_week'])['ride_length'].mean().reset_index(name='avg_duration')
avg_duration_day = avg_duration_day.pivot(index='day_of_week', columns='member_casual', values='avg_duration')
```
